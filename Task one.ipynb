**Objective:** Create a Python program that simulates a game of Rock, Paper, Scissors between the user and the computer.

```python
#import the random library
import random 

#state all posssible choices
choices = ["Rock", "Paper", "Scissors"]

#get the choice of the user
user_name = input("Please Enter Your Name:  ")
user_choice = input("Enter your choice: 'Rock', 'Paper', 'Scissors':  ")
user_choice = user_choice.capitalize()

#verifying if the User choice was part of all possible choices
if user_choice in choices:
    
    #get the computer choice (randomly)
    computer_choice = random.choice(choices)
    #Print the choices
    print('********' * 20)
    print('Rock, Paper, Scissors, Shoot.....')
    print('........' * 20)
    print(f'{user_name}: {user_choice} \
            computer choice:  {computer_choice}')
    print('********' * 20)

    #The Rules of the game
    
    #The Winners
    #'Rock' beats 'Scissors'
    if user_choice == "Rock" and computer_choice == "Scissors":
        print('Yah! You won')

    # 'Scissors' beats 'Paper'
    elif user_choice == 'Scissors' and computer_choice == 'Paper':
        print('Yah! You Won')

    # 'Paper' beats 'Rock'
    elif user_choice == 'Paper' and computer_choice == 'Rock':
        print('Yah! You Won')

    #The Draw
    elif user_choice == computer_choice:
        print('It is a Tie')

    else: 
        print('You lost')
else:
    print(f'Your choice {user_choice} is not part of the choices needed, Kindly select a choice between "Rock", "Paper" and "Scissors". ')
    
    
    
```
