Creating a python program that simulates an AI Agent to learn to play the game TIC-TAC-TOE

```python
import random

#global variables for the game
board = ["-", "-", "-",
         "-", "-", "-",
         "-", "-", "-"]
current_player = "X"
winner = None
gameRunning = True

#create a game board
def PrintBoard(board):
  print("--" * 5)
  print(board[0] + " | " + board[1] + " | " + board[2] )
  print("--" * 5)
  print(board[3] + " | " + board[4] + " | " + board[5] )
  print("--" * 5)
  print(board[6] + " | " + board[7] + " | " + board[8] )
  print("--" * 5)

#Take Player input
def PlayerInput(board):
  try :
    player_input = int(input("Select a number between 1-9: "))

  except ValueError as wrong_value:
    print("Incorrect Value Selected, Ensure you selection is a whole number between 1-9")

  else:
    if player_input in range(1,10) and board[player_input - 1] == "-":
      board[player_input - 1] = current_player

    elif player_input not in range(1,10):
      print("The selected number is out of required range, Kindly select a number between 1-9")

    else:
      print("Ooops The spot has already been selected")





#Check if it is a win or tie

#check win if all spot match horizontally
def  CheckHorizontal(board):
  global winner
  if board[0] == board[1] == board[2] and board[0] != "-":
    winner = board[2]
    return True

  elif board[3] == board[4] == board[5] and board[4] != "-":
    winner = board[5]
    return True

  elif board[6] == board[7] == board[8] and board[8] != "-":
    winner = board[6]
    return True


#winner if all spots match vertically
def CheckVertical(board):
 global winner
 if board[0] == board[3] == board[6] and board[3] != "-":
  winner = board[6]
  return True

 elif board[1] == board[4] == board[7] and board[7] != "-":
  winner = board[1]
  return True

 elif board[2] == board[5] == board[8] and board[2] != "-":
  winner = board[5]
  return True


#winner if all spot match diagonally
def CheckDiag(board):
  global winner

  if board[2] == board[4] == board[6] and board[2] != "-":
    winner = board[2]
    return True
  elif board[0] == board[4] == board[8] and board[6] != "-":
    winner = board[8]
    return True

  #tie if no spot match and space in board are exhausted
def CheckTie(board):
  global gameRunning
  if "-" not in board:
    PrintBoard(board)
    print("It is a Tie")
    gameRunning = False

#check for win
def CheckWin():
  global gameRunning
  if CheckDiag(board) or CheckHorizontal(board) or CheckVertical(board):
    PrintBoard(board)
    print(f"The Winner is {winner}")
    gameRunning = False



#Switch to second player
def SwitchPlayer():
  global current_player
  if current_player == "X":
    current_player = "O"
  else:
    current_player = "X"

#Computer
def computer(board):
  global gameRunning
  while current_player == "O":
    play = random.randint(0,9)
    if board[play] == "-":
      board[play] = "O"
      SwitchPlayer()

      if CheckWin() or CheckTie(board):
        gameRunning = False





#Check if move was a win or tie
```

```python

while gameRunning :
  PrintBoard(board)
  PlayerInput(board)
  CheckWin()
  CheckTie(board)
  SwitchPlayer()
  computer(board)
  CheckWin()
  CheckTie(board)
```
