```python
#load into a pandas dataframe
df = pd.read_csv(os.path.join(path, "amazon_sales_dataset.csv"))
```

After the Data Exploration we discovered that the dataset has 13 columns of which 9 are numerical (4 columns are of float datatype, 5 columns are integer datatype) and 4 columns are categorical/ non-numerical (object datatype).

The column "order_id" appears to be the unique identifier of the dataset of which 5000 non-missing unique observations were recorded.

The dataset appears to be symmetrical and not skewed as the mean is approximately equal to the medians across all columns.

After the Data Exploration we were tasked to to calculate some basic statistics in the dataset.

firstly we calculate the mean, median and standard deviation of the numerical columns

```python
#BASIC STAISTICS OF THE NUMERICAL COLUMN

#selecting the numerical columns
numerical_columns = df.select_dtypes(["int", "float"]).columns
print(f"The numerical columns are : {numerical_columns}")

#calculate the mean, median, standard deviation of the numerical columns
for col in numerical_columns:
  #calculate the mean
  mean = df[col].mean()
  print("\n",f"This is the mean of {col}: {mean}", "\n")

  #calculate the median
  median = df[col].median()
  print(f"The median for the {col} column is {median}", "\n")

  #calculate the standard deviation
  std_dev = df[col].std()
  print(f"The standard Deviation of the {col} column is {std_dev}", "\n")
```

Looking at the mean, median and standard deviation of the numerical columns, it clearly depicts the same figures as gotten in the summary statistics of the dataframe. Next we get the correlation of different features in the dataset.
I would be using a correlation matrix to get the relationship and how strongly correlated the different features of the dataset are to each other.

```python
# create a correlation matrix
correlation_matrix = df.corr(numeric_only=True)

correlation_matrix
```
